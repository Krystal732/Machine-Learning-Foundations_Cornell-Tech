```python
#One hot encode variables of type object
to_encode = ["neighbourhood_group_cleansed", "room_type"]

# Create the encoder:
encoder = OneHotEncoder(handle_unknown="error", sparse=False)

# Apply the encoder:
df_enc = pd.DataFrame(encoder.fit_transform(df[to_encode]))

# Reinstate the original column names:
df_enc.columns = encoder.get_feature_names(to_encode)

df.drop(columns = to_encode ,axis=1, inplace=True)
df = df.join(df_enc)

df = df.select_dtypes(exclude=['object'])
df



```

```python
#remove columns with nan values
nan_count = np.sum(df.isnull(), axis = 0)

condition = nan_count != 0 # look for all columns with missing values

col_names = nan_count[condition].index # get the column names
nan_cols = list(col_names) # convert column names to list
df.drop(columns=nan_cols, axis=1, inplace=True)

```

```python
#Creating features and labels
Y = df['instant_bookable']
X = df.drop(columns = 'instant_bookable', axis=1)

#train test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=123)

#create and fit decision tree model
model = DecisionTreeClassifier(criterion = "entropy", max_depth = 8, min_samples_leaf = 1)
model.fit(X_train, y_train)
class_label_predictions = model.predict(X_test)
acc_score = accuracy_score(y_test, class_label_predictions)
```
